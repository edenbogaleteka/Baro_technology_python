***standard_calculator**

```python
def calculator(): 
    print('standard calculator')
    try:
        while True :
            # begining of operator module
            operator=input('enter the operator [+,-,*,**,/and exit]').strip().lower()
            if operator == 'exit':
                print('exiting the calculator')
                break
            list_op=['+','-','*','/','**']
            if operator not in list_op:
                print('please enter the correct operator')
                continue
                # end of operator module

               # begining of number module
            numbers=[]
            while  True:
                number=input('please enter a number or = sign in order to  calculate').strip()

                if number.lower()=='=':
                    if len(numbers)>2:
                        print('enter atleast 2 numbers')
                        continue
                    break
                try:
                        num=float(number)
                        numbers.append(num)
                except ValueError:
                    print('please enter number only!')
                    continue
                # end of number module

                # begining of arthimetic operation
            
            if operator == '+':
                result =numbers[0]
                for number in numbers[1:]:
                    result=result+number
                print('+'.join(map(str,numbers)) + f" = {result}")
            elif operator == '-':
                result =numbers[0]
                for number in numbers[1:]:
                    result=result-number
                print('-'.join(map(str,numbers)) + f" = {result}")
            elif operator == '*':
                result =numbers[0]
                for number in numbers[1:]:
                    result=result*number
                print('*'.join(map(str,numbers)) + f" = {result}")
            elif operator == '/':
                result =numbers[0]
                try:
                    for number in numbers[1:]:
                        if number == 0:
                            raise ZeroDivisionError()
                    result=result/number
                    print('/'.join(map(str,numbers)) + f" = {result}")
                except ZeroDivisionError:
                    print('denominator can not be zero' )
            elif operator == '**':
                result =numbers[0]
                for number in numbers[1:]:
                    result=result**number
                print('**'.join(map(str,numbers)) + f" = {result}")
    except Exception as e:
        print(f'invalid operations!{e} ')
    print('developed by: EDEN')
    
calculator()   
```
